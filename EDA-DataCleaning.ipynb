**üîÅ Step 2: Data Cleaning & Type Conversion**

</small>

```python
# Clean and Convert Price and Rating Columns
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ','')
df['discounted_price'] = df['discounted_price'].str.replace(',','')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce') #errors='coerce' forces invalid entries (e.g., "N/A") to become NaN.

df['actual_price'] = df['actual_price'].str.replace('‚Çπ','')
df['actual_price'] = df['actual_price'].str.replace(',','')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce') / 100

df['rating'] = df['rating'].str.replace(',','')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce').astype('Int64')
```

```python
#Check Data types
df.dtypes
```

<small>

**üßπ Step 3: Handling Missing Values**

</small>

```python
df.isnull().sum()
```

```python
# Manual fixes based on external lookups 
df.loc[df['product_id'] == 'B08L12N5H1', 'rating'] = 3.9
df.loc[df['product_id'] == 'B0B94JPY2N', 'rating_count'] = 2201

# Drop rows where rating_count is still missing
df = df.dropna(subset=['rating_count'])
```

```python
# Summary statistics
df.describe()
```

```python
 #Visualizing Distributions
def plot_box(column):
    plt.figure(figsize=(8, 1.5))
    sns.boxplot(x=column, data=df, color='skyblue')
    plt.title(f'Boxplot of {column}')
    plt.show()

plot_box('actual_price')
plot_box('discounted_price')
plot_box('discount_percentage')
plot_box('rating_count')
plot_box('rating')
```

<small>

**‚≠ê Step 6: Bayesian Rating**

</small>

```python
# Weighted Rating Calculation (Bayesian Rating)
C = df['rating'].mean()
m = 100  # Minimum votes to be considered

df['bayesian_rating'] = (
    (df['rating_count'] / (df['rating_count'] + m)) * df['rating'] +
    (m / (df['rating_count'] + m)) * C
)
```

<small>

**üóÇ Step 7: Expand Category Column**

</small>

```python
# Split category into subcategories
category_splits = df['category'].str.split('|', expand=True)
for i in range(category_splits.shape[1]):
    df[f'subcategory_{i+1}'] = category_splits[i]

df[['category'] + [f'subcategory_{i+1}' for i in range(category_splits.shape[1])]].head()
```
